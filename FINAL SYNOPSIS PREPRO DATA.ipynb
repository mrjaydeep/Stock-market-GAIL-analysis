```python
# Importing libraries and using pandas library importing dataset

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from datetime import datetime


data= pd.read_csv('GAIL.csv')
print(data)
print(data.dtypes)
```

```python
#printing out the null values from the data set
print(data[['Trades',"Deliverable Volume",'%Deliverble']].isnull())
```

```python
#replacing the null values with zero
print ("NaN replaced with '0':")
newdataset= data.fillna(0)
print(newdataset)
```

```python
#using the describe method describing the data
data.describe()
```

```python
#Starting preprocessing
newdata=newdataset.drop(labels=['Date','Symbol','Series','Turnover','%Deliverble'],axis=1)
print(newdata)
```

```python
#using normalization, normalizing the column using preprocessing
from sklearn import preprocessing
x_array = np.array(data['Prev Close'])
new = preprocessing.normalize([x_array])
b=x_array.reshape(-1,1)

print(b)
```

```python
#finding the avg daily return
newdataset['Daily Return'] = newdataset['Close'].pct_change()
avg_daily_return = newdataset['Daily Return'].mean()
print(f"The average daily return of GAIL during the period 2000-2021 was {avg_daily_return:.4f}.")
```

```python
#finding the correlation 
corr_coef = newdata['Close'].corr(newdata['Volume'])
#If the coefficient is close to 1, it indicates a strong positive correlation, while a coefficient close to -1 indicates a strong negative correlation. A coefficient close to 0 indicates no correlation.
print(f"The correlation coefficient between the GAIL Company Closing value and the trading volume is {corr_coef:.4f}.")
```

```python
#finding average daily return by year 
newdataset['Year'] = pd.DatetimeIndex(newdataset['Date']).year
avg_daily_returns_by_year = newdataset.groupby('Year')['Daily Return'].mean()
print(avg_daily_returns_by_year)
```

```python
#plotting graph of avg_daily_return by year graph
plt.bar(avg_daily_returns_by_year.index, avg_daily_returns_by_year,color='r')
plt.xlabel('Year')
plt.ylabel('Average Daily Return')
plt.title('Average Daily Return by Year')
plt.show()
```
